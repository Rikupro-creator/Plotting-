```python
import numpy as np
import matplotlib.pyplot as plt

# Dose levels and corresponding effects for a single drug
dose_levels_single = np.array([0, 1, 2, 3, 4, 5])
effect_single = np.array([0, 10, 30, 60, 90, 100])

# Dose levels and corresponding effects for a drug with biphasic effect
dose_levels_biphasic = np.array([0, 1, 2, 3, 4, 5])
effect_biphasic = np.array([0, 20, 70, 90, 60, 100])

# Create a subplot for single drug dose-response curve
plt.subplot(2, 1, 1)
plt.plot(dose_levels_single, effect_single, marker='o', linestyle='-', color='blue')
plt.xlabel('Dose (mg)')
plt.ylabel('Effect (%)')
plt.title('Dose-Effect Curve for a Single Drug')
plt.grid(True)

# Create a subplot for biphasic drug dose-response curve
plt.subplot(2, 1, 2)
plt.plot(dose_levels_biphasic, effect_biphasic, marker='o', linestyle='-', color='green')
plt.xlabel('Dose (mg)')
plt.ylabel('Effect (%)')
plt.title('Dose-Effect Curve for a Drug with Biphasic Effect')
plt.grid(True)

plt.tight_layout()  # Ensures subplots don't overlap
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Logistic sigmoid function
def sigmoid(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))

# Dose levels and corresponding effects for Drug A with a steep slope
dose_levels_a = np.array([0, 1, 2, 3, 4, 5])
effect_a = np.array([0, 20, 70, 90, 95, 100])

# Dose levels and corresponding effects for Drug B with a gradual slope
dose_levels_b = np.array([0, 1, 2, 3, 4, 5])
effect_b = np.array([0, 10, 30, 60, 80, 100])

# Fit sigmoid curves to the data for Drug A
params_a, _ = curve_fit(sigmoid, dose_levels_a, effect_a, maxfev=10000)

# Fit sigmoid curves to the data for Drug B
params_b, _ = curve_fit(sigmoid, dose_levels_b, effect_b, maxfev=10000)

# Generate smooth dose values for the curves
smooth_dose_levels = np.linspace(0, 5, 100)

# Calculate the corresponding effects using the fitted sigmoid functions
smooth_effect_a = sigmoid(smooth_dose_levels, *params_a)
smooth_effect_b = sigmoid(smooth_dose_levels, *params_b)

# Create the dose-effect curves with sigmoid fits
plt.plot(smooth_dose_levels, smooth_effect_a, color='blue', label='Drug A (Steep Slope)')
plt.plot(smooth_dose_levels, smooth_effect_b, color='green', label='Drug B (Gradual Slope)')

plt.xlabel('Dose (mg)')
plt.ylabel('Effect (%)')
plt.title('S-Shaped Dose-Effect Curves for Two Hypothetical Drugs')
plt.grid(True)
plt.legend()

plt.show()
```
