## making figures with numpy array

We want:
* line plot

```python
# create some data
x=np.linspace(0,10,  100)
```

```python
# plot the data

fig, ax=plt.subplots()
ax.plot(x, x**2);
```

```python
# make a plot from a dictionary
nut_butter_prices= {"almond butter":10,
                    'peanut butter':8,
                    'casher butter':12}
fig, ax=plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="nut butter prices",
       ylabel='Total $',
       xlabel='Butter types')
```

```python
# horizontal bar graphs
fig, ax= plt.subplots()
ax=plt.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# make some data for histogram and plot it
x=np.random.randn(1000)
fig, ax=plt.subplots()
ax.hist(x)
ax.set(title='Histogram of normal datasets');
```

## two options for subplots

```python
fig, ((ax1, ax2), (ax3, ax4))=plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,7))
# plot for each of the axes
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.random(100));
```

####Option 2

```python
fig, ax=plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,10))
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(100));
```

### which one should you use (pyplot vs matplot oo method)
*when plotting someting quickly its okay to use the pyplot method
* when plotting something more advanced its okay to use the oo method

```python
over_50=hearts[hearts['age']>50]
over_50
```

```python
len(over_50)
```

```python
#pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target');
```

```python
## oo method
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age', y='chol', c ='target', ax=ax)
#ax.set_xlim([45, 100]);
```

```python
# opp method from scratch
fig, ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])
ax.set(title='Heart disease and cholestorel levels', xlabel='age', ylabel='cholesterol leves')
ax.legend(*scatter.legend_elements(), title='Target')
# add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
# subplots of age, chol, thalach
fig, (ax0, ax1)= plt.subplots(
                             nrows=2,
                             ncols=1,
                             figsize=(10, 10),
                             sharex=True
                        )

scatter=ax0.scatter(x=over_50['age'], y=over_50['chol'],
                    c=over_50['target'])
ax0.set(title='heart diseases vs cholestorol',
        ylabel='Cholestorol')
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean())
scatter2=ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])
ax1.set(title='age vs thalach', xlabel='Age', ylabel='Thalach')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean(), linestyle='dotted')

# add a title to our figure
fig.suptitle('Heart Disease analysis', fontsize=16, fontweight='bold');
```
