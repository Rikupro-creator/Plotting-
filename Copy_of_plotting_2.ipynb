```python
import matplotlib.pyplot as plt

def demand_curve(quantity_demanded):
    return 100 - 2 * quantity_demanded

def monopolist_revenue(quantity_sold):
    return quantity_sold * demand_curve(quantity_sold)

# Simulate and plot the demand curve
quantity_values = list(range(0, 51))
price_values = [demand_curve(quantity) for quantity in quantity_values]

plt.plot(quantity_values, price_values)
plt.xlabel("Quantity Sold")
plt.ylabel("Price")
plt.title("Monopolist Demand Curve")
plt.grid(True)
plt.show()

# Simulate and plot the total revenue curve for the monopolist
revenue_values = [monopolist_revenue(quantity) for quantity in quantity_values]

plt.plot(quantity_values, revenue_values)
plt.xlabel("Quantity Sold")
plt.ylabel("Total Revenue")
plt.title("Monopolist Total Revenue Curve")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def demand_curve(quantity_demanded):
    return 100 - 2 * quantity_demanded

def marginal_revenue(quantity_sold):
    return 100 - 4 * quantity_sold

# Simulate and plot the demand curve
quantity_values = list(range(0, 51))
price_values = [demand_curve(quantity) for quantity in quantity_values]

plt.plot(quantity_values, price_values, label="Demand Curve")
plt.xlabel("Quantity Sold")
plt.ylabel("Price")
plt.title("Monopolist Demand Curve and Marginal Revenue")
plt.grid(True)

# Simulate and plot the marginal revenue curve for the monopolist
marginal_revenue_values = [marginal_revenue(quantity) for quantity in quantity_values]

plt.plot(quantity_values, marginal_revenue_values, label="Marginal Revenue")
plt.legend()
plt.show()
```
