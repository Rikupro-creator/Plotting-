```python
import numpy as np
import matplotlib.pyplot as plt

# Define the fixed costs and production quantity
fixed_costs = 1000  # Total fixed costs
production = np.arange(100, 1001, 100)  # Varying production quantity

# Calculate average fixed cost (AFC)
AFC = fixed_costs / production

# Plotting AFC against production
plt.plot(production, AFC)
plt.xlabel('Production Quantity')
plt.ylabel('Average Fixed Cost')
plt.title('Relationship between Production Quantity and Average Fixed Cost')
plt.grid()
plt.show()
```

```python
# Define the variable costs and production quantity
variable_costs = 500  # Total variable costs
production = np.arange(100, 1001, 100)  # Varying production quantity

# Calculate average variable cost (AVC)
AVC = variable_costs / production

# Plotting AVC against production
plt.plot(production, AVC)
plt.xlabel('Production Quantity')
plt.ylabel('Average Variable Cost')
plt.title('AVC vs. Production Quantity')
plt.show()
```

```python
# Define the fixed costs, variable costs, and production quantity
fixed_costs = 1000  # Total fixed costs
variable_costs = 500  # Total variable costs
production = np.arange(100, 1001, 100)  # Varying production quantity

# Calculate average total cost (ATC)
total_costs = fixed_costs + variable_costs
ATC = total_costs / production

# Plotting ATC against production
plt.plot(production, ATC)
plt.xlabel('Production Quantity')
plt.ylabel('Average Total Cost')
plt.title('ATC vs. Production Quantity')
plt.show()
```

```python
# Define the fixed costs, variable costs, and production quantity
fixed_costs = 1000  # Total fixed costs
variable_costs = 500  # Total variable costs
production = np.arange(100, 1001, 100)  # Varying production quantity

# Calculate average total cost (ATC)
total_costs = fixed_costs + variable_costs * production
ATC = total_costs / production

# Reverse the ATC values to create an increasing curve
ATC = np.flip(ATC)

# Plotting ATC against production
plt.plot(production, ATC)
plt.xlabel('Production Quantity')
plt.ylabel('Average Total Cost')
plt.title('ATC vs. Production Quantity')
plt.show()
```
